```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv( 'adult.csv' )
data.head(10)
```

```python
data.info()
```

```python
print('Для educational-num пустых строк {}'.format( len( data[ data['educational-num'].isna() ] ) ))
```

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit( data['income'] )
```

```python
le.classes_
```

```python
y = pd.Series( data = le.transform( data['income'] ) )
y.head()
```

```python
# попробуем выбрать какие-нибудь признаки
selectedColumns = data[ [ 'educational-num', 'occupation', 'race', 'gender', 'capital-gain', 'hours-per-week',  'income' ] ]
```

```python
X = pd.get_dummies( selectedColumns, columns = [ 'occupation', 'race', 'gender'  ] )
```

```python
del X['income']
X.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# обучаем модель

model.fit( X, y )
predictions = model.predict_proba( X )
```

```python
model.score(X, y)
```
